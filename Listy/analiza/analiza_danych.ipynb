```python
import numpy as np
import matplotlib.pyplot as plt
import os

# indexy: dodawanie pocztaek (0), dodawanie na koniec (1), dodawanie na losowy indeks (2)
# indexy: usuwanie poczatek (3), usuwanie z konca (4), usuwanie z losowego indeksu (5)
# indexy: wyszukiwanie indeksu danego elementu (6)
# format w pliku (csv): {index testu},{czas (ns)} 
# nazwa pliku csv to {n elementow}_{lista / tablica}.csv
# format w tablicy koncowych wynikow do wykresow 
# ["lista/tablica", (n elementow), (czas sredni), (rozpietosc q3 - q1)]


# w tych tablicach jest format 
# ["struktura (lista / tablica)", (n elementow), czas]
test_0 = []
test_1 = []
test_2 = []
test_3 = []
test_4 = []
test_5 = []
test_6 = []


files = os.listdir("./wyniki")
```

```python
print(files)
```

```python
for filename in files:
    liczba_el = int(filename.split('_')[0])
    struktura = filename.split('_')[1].split('.')[0]
    linie_pliku = []

    with open(f"./wyniki/{filename}") as dataFile:
        linie_pliku = dataFile.readlines()

    for linia in linie_pliku:
        linia = linia.split(',')
        test = int(linia[0])
        gotowy_wiersz = [struktura, liczba_el, int(linia[1])]

        if(test == 0):
            test_0.append(gotowy_wiersz)
        elif(test == 1):
            test_1.append(gotowy_wiersz)
        elif(test == 2):
            test_2.append(gotowy_wiersz)
        elif(test == 3):
            test_3.append(gotowy_wiersz)
        elif(test == 4):
            test_4.append(gotowy_wiersz)
        elif(test == 5):
            test_5.append(gotowy_wiersz)
        elif(test == 6):
            test_6.append(gotowy_wiersz)
```

```python
for i in test_0:
    print(i)
```

```python
# odrzucenie wsztkiego z poza przedzialy (kwantyl1, kwantyl3)

def odrzuc_skrajne_dane(arr: list, liczba_el: int) -> list:
    result = []
    wyniki_same_tabela = []
    wyniki_same_lista = []
    for i in arr:
        if i[0] == 'tablica' and i[1] == liczba_el:
            wyniki_same_tabela.append(i[2])
        elif i[0] == 'lista' and i[1] == liczba_el:
            wyniki_same_lista.append(i[2])

    q3_t, q1_t = np.quantile(wyniki_same_tabela, [0.75, 0.25])
    q3_l, q1_l = np.quantile(wyniki_same_lista, [0.75, 0.25])
    print(f"liczba elementow: {liczba_el}")
    print(f"q1_t: {q1_t} q3_t: {q3_t}")
    print(f"q1_l: {q1_l} q3_l: {q3_l}")

    for i in arr:
        if(i[2] > q1_t) and (i[2] < q3_t) and i[0] == "tablica" and i[1] == liczba_el:
            result.append(i)
        elif(i[2] > q1_l) and (i[2] < q3_l) and i[0] == "lista" and i[1] == liczba_el:
            result.append(i)

    return result

def oblicz_srednia(arr: list, liczba_el: int)->list:
    wyniki_same_tabela = []
    wyniki_same_lista = []
    for i in arr:
        if i[0] == 'tablica' and i[1] == liczba_el:
            wyniki_same_tabela.append(i[2])
        elif i[0] == 'lista' and i[1] == liczba_el:
            wyniki_same_lista.append(i[2])

    return [
        ['tablica', round(float(np.mean(wyniki_same_tabela)), 2)],
        ['lista', round(float(np.mean(wyniki_same_lista)), 2)]
    ]

for x in oblicz_srednia(odrzuc_skrajne_dane(test_0, 10000), 10000):
    print(x)
```

```python
def stworz_wykres(test_arr: list, title: str):
    x = [10000, 20000, 30000, 40000, 50000]

    k_10 = oblicz_srednia(odrzuc_skrajne_dane(test_arr, 10000), 10000)
    k_20 = oblicz_srednia(odrzuc_skrajne_dane(test_arr, 20000), 20000)
    k_30 = oblicz_srednia(odrzuc_skrajne_dane(test_arr, 30000), 30000)
    k_40 = oblicz_srednia(odrzuc_skrajne_dane(test_arr, 40000), 40000)
    k_50 = oblicz_srednia(odrzuc_skrajne_dane(test_arr, 50000), 50000)

    y_tablica = [k_10[0][1], k_20[0][1], k_30[0][1], k_40[0][1], k_50[0][1]]
    y_lista = [k_10[1][1], k_20[1][1], k_30[1][1], k_40[1][1], k_50[1][1]]

    plt.figure(figsize=(10,5))
        # Tworzenie wykresu
    plt.plot(x, y_tablica, marker='x', label="Tablica")
    for i in range(len(x)):
        plt.annotate(str(y_tablica[i]), (x[i], y_tablica[i]), textcoords="offset points", xytext=(0, 10), ha='center')
    plt.plot(x, y_lista, marker='x', label="Lista")
    for i in range(len(x)):
        plt.annotate(str(y_lista[i]), (x[i], y_lista[i]), textcoords="offset points", xytext=(0, 10), ha='center')
    # Dodawanie tytułu i etykiet osi (opcjonalnie, ale to dobra praktyka)
    plt.title(title)
    plt.xlabel("Liczba elementow")
    plt.ylabel("Czas operacji (ns)")
    plt.xlim(10000, 50000)
    plt.legend()
    # Wyświetlenie wykresu
    plt.show()
```

```python
stworz_wykres(test_0, "Dodawanie na pierwszy indeks")
```

```python
stworz_wykres(test_1, "Dodawanie na koniec")
```

```python
stworz_wykres(test_2, "Dodawanie na losowy indeks")
```

```python
stworz_wykres(test_3, "Usuwanie z pierwszego indeksu")
```

```python
stworz_wykres(test_4, "Usuwanie z ostatniego indeksu")
```
