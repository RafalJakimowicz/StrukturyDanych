```python
import numpy as np
import matplotlib.pyplot as plt
import os

# indexy: dodawanie pocztaek (0), dodawanie na koniec (1), dodawanie na losowy indeks (2)
# indexy: usuwanie poczatek (3), usuwanie z konca (4), usuwanie z losowego indeksu (5)
# indexy: wyszukiwanie indeksu danego elementu (6)
# format w pliku (csv): {index testu},{czas (ns)} 
# nazwa pliku csv to {n elementow}_{lista / tablica}.csv
# format w tablicy koncowych wynikow do wykresow 
# ["lista/tablica", (n elementow), (czas sredni), (rozpietosc q3 - q1)]


# w tych tablicach jest format 
# ["struktura (lista / tablica)", (n elementow), czas]
test_0 = []
test_1 = []
test_2 = []
test_3 = []
test_4 = []
test_5 = []
test_6 = []


files = os.listdir("./wyniki")
```

```python
print(files)
```

```python
for filename in files:
    liczba_el = int(filename.split('_')[0])
    struktura = filename.split('_')[1].split('.')[0]
    linie_pliku = []

    with open(f"./wyniki/{filename}") as dataFile:
        linie_pliku = dataFile.readlines()

    for linia in linie_pliku:
        linia = linia.split(',')
        test = int(linia[0])
        gotowy_wiersz = [struktura, liczba_el, int(linia[1])]

        if(test == 0):
            test_0.append(gotowy_wiersz)
        elif(test == 1):
            test_1.append(gotowy_wiersz)
        elif(test == 2):
            test_2.append(gotowy_wiersz)
        elif(test == 3):
            test_3.append(gotowy_wiersz)
        elif(test == 4):
            test_4.append(gotowy_wiersz)
        elif(test == 5):
            test_5.append(gotowy_wiersz)
        elif(test == 6):
            test_6.append(gotowy_wiersz)
```
